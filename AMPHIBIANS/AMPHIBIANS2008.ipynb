```python
# Read 2019 data for amphibians, mammals and reptiles

Amphibians_2019 = pd.read_csv("ANIMALS.csv")
```

```python
# Calculate the frequency of the English names for 2008
 
Amphibians_2008['freq'] = Amphibians_2008.groupby('Vernacular name - original')['Vernacular name - original'].transform('count')

# Drop rows with English name duplicates in 2008

Amphibians_2008.drop_duplicates(subset ="Vernacular name - original", keep = "first", inplace = True) 
```

```python
# Check the first 5 rows in 2008

Amphibians_2008.head()
```

```python
# Add the frequency in 2008 to 2019

for index, row in Amphibians_2019.iterrows():
    species = row['Vernacular name - original']
    if any(Amphibians_2008['Vernacular name - original'] == species):
        Amphibians_2019.at[index,'2008'] = Amphibians_2008.loc[Amphibians_2008['Vernacular name - original'] == species, 'freq'].iloc[0]
        
```
