```python
# Calculate the frequency of the English names for 2019

Mammals_2019['freq'] = Mammals_2019.groupby('Vernacular name - original')['Vernacular name - original'].transform('count')

# Take the rows where the frequency is less than or equal to 10 in 2019

Mammals_2019 = Mammals_2019[Mammals_2019['freq'] <= 10]
Mammals_2019.reset_index(inplace=True, drop=True)
Mammals_2019['Count_1968'] = 'NaN'
Mammals_2019['Rate_1968'] = 'NaN'
Mammals_2019.head()
```

```python
# Calculate the frequency of the English names for 1968

Mammals_1968['freq'] = Mammals_1968.groupby('Vernacular name - original')['Vernacular name - original'].transform('count')

# Drop rows with English name duplicates in 1968

Mammals_1968.drop_duplicates(subset ="Vernacular name - original", keep = "first", inplace = True) 
```

```python
# Add the frequency in 1968 to 2019

for index, row in Mammals_2019.iterrows():
    species = row['Vernacular name - original']
    if any(Mammals_1968['Vernacular name - original'] == species):
        Mammals_2019.at[index,'Count_1968'] = Mammals_1968.loc[Mammals_1968['Vernacular name - original'] == species, 'freq'].iloc[0]
        
    else:
        Mammals_2019.at[index,'Count_1968'] = 0
        
```

```python
# Calculate the rate of extinction

for index, row in Mammals_2019.iterrows():
    freq_2019 = row['freq']
    freq_1968 = row['Count_1968']
    
    if freq_1968 != 0:
        rate = ((freq_1968 - freq_2019)/freq_1968)*100
        
        if rate > 0:
            Mammals_2019.at[index,'Rate_1968'] = rate * -1
        else:
            Mammals_2019.at[index,'Rate_1968'] = 0
    else:
        Mammals_2019.at[index,'Rate_1968'] = 0
        
```

```python
# Take the rows where the rate is greater than or equal to 50%

Mammals_2019 = Mammals_2019[Mammals_2019['Rate_1968'] <= -50]

# Check the first 5 rows in 2019

Mammals_2019.head()
```
